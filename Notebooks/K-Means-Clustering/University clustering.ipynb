## 5. K Means Cluster Creation

### 5.1 Importing the libraries
**Import KMeans from SciKit Learn.**

```python
from sklearn.cluster import KMeans
```

### 5.2 Training the model

**Creating an instance of a K-Means model with 2 clusters.**

```python
kmeans = KMeans(n_clusters=2)
```

**Fitting the model to all the data, except for the Private label.**

```python
kmeans.fit(data.drop(["Private"], axis=1))
```

----
**Cluster Center Vectors**

-----

```python
kmeans.cluster_centers_
```

## 6. Evaluation 

There is no perfect way to evaluate clustering if you don't have the labels. However since we do have the labels, so we take advantage of this to evaluate our clusters.

**Create a new column for df called 'Cluster', which is a 1 for a Private school, and a 0 for a public school.**

```python
data["Cluster"] = kmeans.labels_
```

```python
data.head()
```

**Creating a confusion matrix and classification report to see how well the Kmeans clustering worked without being given any labels.**

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
y_test = data['Private'].apply(lambda x : 1 if x == "Yes" else 0)
print(confusion_matrix(y_test, data['Cluster']))
print(classification_report(y_test, data['Cluster']))
```
